```python
import requests
from bs4 import BeautifulSoup

# URL de la página que quieres scrapear
url = 'https://www.inmobiliariabogota.com/local.php'

# Realiza la solicitud HTTP a la página
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Encuentra los contenedores de los anuncios
listings = soup.find_all('div', class_='ulisting-item-list')

# Itera sobre cada contenedor de anuncio
for listing in listings:
    try:
        # Extraer el título del anuncio
        title = listing.find('a', class_='attribute-title-box').text.strip()

        # Extraer el precio (basado en la clase genuine_sale)
        price_element = listing.find('div', class_='genuine_sale')
        price = price_element.text.strip() if price_element else 'No disponible'

        # Extraer el área
        area_element = listing.find('div', class_='attribute-value')
        area = area_element.text.strip() if area_element and 'm²' in area_element.text else 'No disponible'

        # Extraer la descripción (dentro de ulisting-element)
        description_element = listing.find('div', class_='ulisting-element')
        description = description_element.text.strip() if description_element else 'No disponible'

        # Extraer los atributos como garajes, baños, habitaciones
        attributes = listing.find_all('div', class_='attribute-parts-wrap')
        for attribute in attributes:
            attribute_name = attribute.find('div', class_='attribute-name').text.strip()
            attribute_value = attribute.find('div', class_='attribute-value').text.strip()
            print(f'{attribute_name}: {attribute_value}')
        
        # Imprimir los resultados extraídos
        print(f'Título: {title}')
        print(f'Precio: {price}')
        print(f'Área: {area}')
        print(f'Descripción: {description}')
        print('-' * 40)

    except AttributeError:
        # Maneja cualquier error en la extracción
        continue
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Configuración del controlador de Selenium
driver = webdriver.Chrome()

# URL de la página con los listados
url = 'https://www.inmobiliariabogota.com/local.php'
driver.get(url)

# Esperar a que la página cargue completamente
time.sleep(3)

# Extraer todos los elementos con listados de propiedades
listings = driver.find_elements(By.CSS_SELECTOR, '.ulisting-item-list')

# Iterar sobre cada propiedad
for listing in listings:
    try:
        # Extraer la URL para obtener la ubicación del mapa
        details_url = listing.find_element(By.TAG_NAME, 'a').get_attribute('href')

        # Ir a la página de detalles de la propiedad
        driver.get(details_url)
        time.sleep(3)

        # Buscar el iframe del mapa en la página de detalles
        iframe = driver.find_element(By.CSS_SELECTOR, 'iframe[src*="openstreetmap.org"]')
        map_url = iframe.get_attribute('src')

        # Extraer las coordenadas del URL del iframe
        if 'bbox=' in map_url:
            bbox_params = map_url.split('bbox=')[1].split('%2C')
            longitude = bbox_params[0]
            latitude = bbox_params[1]
            print(f"Coordenadas encontradas: Latitud = {latitude}, Longitud = {longitude}")
        else:
            print("No se encontró mapa con coordenadas en este inmueble.")

        print("-" * 40)

        # Volver a la página principal para seguir con otros listados
        driver.back()
        time.sleep(2)

    except Exception as e:
        print(f"Error al extraer la ubicación: {e}")
        continue

# Cerrar el navegador
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# Configuración de opciones de Selenium para que funcione en segundo plano
chrome_options = Options()
chrome_options.add_argument('--headless')  # Ejecuta en segundo plano sin mostrar la ventana
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Inicializa el driver de Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# URL de la página principal de la primera pestaña
url_base = 'https://www.inmobiliariabogota.com/local.php'

# Crear el archivo CSV y escribir los encabezados
with open('inmuebles_todas_paginas.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Encabezados del CSV
    writer.writerow(['Título', 'Precio', 'Área', 'Descripción', 'Garajes', 'Baños', 'Habitaciones', 'Latitud', 'Longitud'])

    # Iterar sobre las 15 páginas
    for page_num in range(1, 16):
        print(f"Extrayendo datos de la página {page_num}...")
        
        # Modificar la URL para acceder a cada página (si hay una lógica de paginación en la URL, ajusta aquí)
        if page_num > 1:
            url = f'{url_base}?page={page_num}'  # Ajusta la URL según la lógica de paginación de la página
        else:
            url = url_base

        # Cargar la página
        driver.get(url)
        time.sleep(3)

        # Encuentra los contenedores de los anuncios
        listings = driver.find_elements(By.CLASS_NAME, 'ulisting-item-list')

        # Itera sobre cada contenedor de anuncio
        for listing in listings:
            try:
                # Extraer el título del anuncio
                title = listing.find_element(By.CLASS_NAME, 'attribute-title-box').text.strip()

                # Extraer el precio (basado en la clase genuine_sale)
                try:
                    price = listing.find_element(By.CLASS_NAME, 'genuine_sale').text.strip()
                except:
                    price = 'No disponible'

                # Extraer el área directamente desde el elemento de atributo del área
                try:
                    area_element = listing.find_element(By.XPATH, ".//div[contains(text(),'m²')]")
                    area = area_element.text.strip()
                except:
                    area = 'No disponible'

                # Extraer los atributos como garajes, baños, habitaciones
                attributes = listing.find_elements(By.CLASS_NAME, 'attribute-parts-wrap')
                attributes_dict = {}
                for attribute in attributes:
                    attribute_name = attribute.find_element(By.CLASS_NAME, 'attribute-name').text.strip()
                    attribute_value = attribute.find_element(By.CLASS_NAME, 'attribute-value').text.strip()
                    attributes_dict[attribute_name] = attribute_value

                # Valores predeterminados para los atributos
                garajes = attributes_dict.get('Garajes', 'No disponible')
                banos = attributes_dict.get('Baños', 'No disponible')
                habitaciones = attributes_dict.get('Habitaciones', 'No disponible')

                # Extraer la descripción
                try:
                    description = listing.find_element(By.CLASS_NAME, 'ulisting-attribute-box ulisting_element_590').text.strip()
                except:
                    description = 'No disponible'

                # Extraer la URL del detalle del inmueble para obtener la ubicación
                details_url = listing.find_element(By.TAG_NAME, 'a').get_attribute('href')

                # Ir a la página de detalles del inmueble
                driver.get(details_url)
                time.sleep(3)

                # Buscar el iframe del mapa en la página de detalles
                try:
                    iframe = driver.find_element(By.CSS_SELECTOR, 'iframe[src*="openstreetmap.org"]')
                    map_url = iframe.get_attribute('src')

                    # Extraer las coordenadas del URL del iframe
                    if 'bbox=' in map_url:
                        bbox_params = map_url.split('bbox=')[1].split('%2C')
                        longitude = bbox_params[0]
                        latitude = bbox_params[1]
                    else:
                        latitude = 'No disponible'
                        longitude = 'No disponible'
                except:
                    latitude = 'No disponible'
                    longitude = 'No disponible'

                # Escribir los datos en el archivo CSV
                writer.writerow([title, price, area, description, garajes, banos, habitaciones, latitude, longitude])

                # Volver a la página principal para seguir con otros listados
                driver.back()
                time.sleep(2)

            except Exception as e:
                print(f"Error: {e}")
                continue

# Cerrar el navegador
driver.quit()

print("Datos guardados exitosamente en 'inmuebles_todas_paginas.csv'")
```
